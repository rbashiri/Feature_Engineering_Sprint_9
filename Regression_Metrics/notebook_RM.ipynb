The data we preprocessed previously has been saved to /datasets/flights_preprocessed.csv. In the precode it is now uploaded from the file and stored in the predicted_valid variable.

You need to:

```python
# spliting data and creating Features and targets 
train, valid = train_test_split(data_ohe, test_size = 0.25, random_state =12345)
X_train = train.drop('Arrival Delay', axis =1)
X_valid = valid.drop( 'Arrival Delay', axis =1)
y_train = train['Arrival Delay']
y_valid = valid['Arrival Delay']
# model and fitting 
model= LinearRegression()
model.fit(X_train, y_train)
predicted_valid = model.predict(X_valid)
mse= mean_squared_error(y_valid, predicted_valid)
print("MSE:", mse)
print()
print('='*45)
print("RMSE:", mse**0.5)


```
