### Mean Absolute Error
  #### Task 6 : 
  
Code the mae() function based on the formula. It takes the actual targets and the predicted targets and returns the mean absolute error value.

`*Note that in Python, absolute value is calculated using the abs() function.*`

Test the mae() function on the example in precode. Print the result

```python

target = pd.Series([-0.5, 2.1, 1.5, 0.3])
predictions = pd.Series([-0.6, 1.7, 1.6, 0.2])

# defines the function 
def mae(target, predictions):
    error = 0 # count 
   # Iterate through each position
    for i in range(target.shape[0]):
        # Accumulate: add absolute difference
        error += (abs(target[i] - predictions[i]))
    
    # Finalize: return average
    return error/len(target)
print(mae(target, predictions))

```
